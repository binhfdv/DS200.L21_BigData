### Datetime normalization

```python
# Normalize Field_34, ngaySinh
def ngaysinh_34_normalize(s):
  if s != s: return np.nan
  try:
    s = int(s)
  except:
    try:
      t = str(datetime.strptime(str(s)[:6], "%Y%m"))
      return t
    except:
      return np.nan

  try:
    t =  str(datetime.strptime(str(s)[:6], "%Y%m"))
    return t
  except:
    return np.nan

spark.udf.register('func1',ngaysinh_34_normalize)
```

```python
def datetime_normalize(s):
  if (s!=s) or (s=='nan') or (s=='DN'): return np.nan #(np.nan!=np.nan) = True
  try:
    s = s.split(".")[0]
  except:
    return np.nan
  if s[-1] == "Z":
    s = s[:-1]
  
  try:
    t = str(datetime.strptime(s, "%Y-%m-%dT%H:%M:%S"))
    return t
  except:
    return np.nan
  
spark.udf.register('func2', datetime_normalize)
```

```python
# Normalize date data
def date_normalize(s):
  if s!=s or s==None or s=='nan': return np.nan
  try:
    t = datetime.strptime(s, "%m/%d/%Y")
  except: 
    try:
      t = datetime.strptime(s, "%Y-%m-%d")
    except:
      return np.nan
  return str(t)

spark.udf.register('func3', date_normalize)
```

```python
def add_delta_cols(s):
  if s==None:
    return str(0)
  t = s.split()
  if len(t)==6:
    return str(int(t[0])*3600+int(t[2])*60+int(t[4]))
  elif len(t)==4:
    return str(int(t[0])*3600+int(t[2])*60)
  elif len(t)==2:
    return str(int(t[0])*3600)

spark.udf.register('func4', add_delta_cols)
```

```python
def add_delta_cols_byday(s):
  if s==None:
    return str(0)
  t = s.split()
  if '0' in t:
    return str(0)
  return str(int(int(t[0])/24))

spark.udf.register('func5', add_delta_cols_byday)
```

```python
date_cols = ["Field_{}".format(i) for i in [5, 6, 7, 8, 9, 11, 15, 25, 32, 33, 35, 40]]
datetime_cols = ["Field_{}".format(i) for i in [1, 2, 43, 44]]
correct_dt_cols = ['Field_34', 'ngaySinh']
cat_cols = date_cols + datetime_cols + correct_dt_cols
cat_cols = list(set(cat_cols))

def process_datetime_cols(df):
  for col in date_cols:
    df = df.withColumn(col, expr(f"func3({col})"))

  for col in datetime_cols:
    df = df.withColumn(col, expr(f"func2({col})"))

  for col in correct_dt_cols:
    df = df.withColumn(col, expr(f"func1({col})"))
  
  # Some delta columns
  for i, j in zip('43 1 2'.split(), '1 2 44'.split()):
    df = df.withColumn(f'Field_{j}', df[f'Field_{j}'].cast(TimestampType()))
    df = df.withColumn(f'Field_{i}', df[f'Field_{i}'].cast(TimestampType()))
  
    df = df.withColumn(f'DT_{j}_{i}', (df[f'Field_{j}'] - df[f'Field_{i}']))

    df = df.withColumn(f'DT_{j}_{i}', df[f'DT_{j}_{i}'].cast('string'))

    df = df.withColumn(f'DT_{j}_{i}', expr(f"func4(DT_{j}_{i})"))

    df = df.withColumn(f'DT_{j}_{i}', df[f'DT_{j}_{i}'].cast('int'))

  for i, j in zip('5 6 7 33 8 11 9 15 25 6 7 8 9 15 25 2'.split(), '6 34 33 40 11 35 15 25 32 7 8 9 15 25 32 8'.split()):
    df = df.withColumn(f'Field_{j}', df[f'Field_{j}'].cast(TimestampType()))
    df = df.withColumn(f'Field_{i}', df[f'Field_{i}'].cast(TimestampType()))
  
    df = df.withColumn(f'DT_{j}_{i}', (df[f'Field_{j}'] - df[f'Field_{i}']))

    df = df.withColumn(f'DT_{j}_{i}', df[f'DT_{j}_{i}'].cast('string'))

    df = df.withColumn(f'DT_{j}_{i}', expr(f"func5(DT_{j}_{i})"))

    df = df.withColumn(f'DT_{j}_{i}', df[f'DT_{j}_{i}'].cast('int'))

    # Age, month
    df = df.withColumn('ngaySinh', df['ngaySinh'].cast(TimestampType()))
    df = df.withColumn('age', 2021-year('ngaySinh'))
    df = df.withColumn('birth_month', month('ngaySinh'))
    # df = df.withColumn('ngaySinh', df['ngaySinh'].cast(TimestampType()))

    # Days from current time & isWeekday
    for col in cat_cols:
        name = col.split('_')[-1]

        df = df.withColumn(f'is_WD_{name}', dayofweek(col).isin(list(range(1, 8))))
        df = df.withColumn(f'days_from_now_{name}', datediff(current_date(), col))
 
        # df = df.withColumn(col, date_format(col, 'M-y'))

  return df
```

```python
def generate_deltas(df):
  # Delta for x_startDate and x_endDate
  for cat in ['F', 'E', 'C', 'G', 'A']:
    df = df.withColumn(f"{cat}_startDate", expr(f"func3({cat}_startDate)"))
    df = df.withColumn(f"{cat}_endDate", expr(f"func3({cat}_endDate)"))

    df = df.withColumn(f"{cat}_startDate", df[f"{cat}_startDate"].cast(TimestampType()))
    df = df.withColumn(f"{cat}_endDate", df[f"{cat}_endDate"].cast(TimestampType()))

    df = df.withColumn(f"{cat}_start_end", datediff(f"{cat}_endDate", f"{cat}_startDate"))
        
  for i, j in zip('F E C G'.split(), 'E C G A'.split()):
      df = df.withColumn(f"{j}_{i}_startDate", datediff(f"{j}_startDate", f"{i}_startDate"))
      df = df.withColumn(f"{j}_{i}_endDate", datediff(f"{j}_endDate", f"{i}_endDate"))
  
  return df
```

### Precessing categoricals

```python
def str_normalize(s):
  if s!=s:
    return np.nan
  s = str(s).strip().lower()
  s = re.sub(' +', " ", s)

  patt1 = re.compile('(\s*)thành phố(\s*)')
  patt2 = re.compile('(\s*)tỉnh(\s*)')
  patt3 = re.compile('(\s*)city(\s*)')
  patt4 = re.compile('(\s*)province(\s*)')
  s = patt1.sub('', s)
  s = patt2.sub('', s)
  s = patt3.sub('', s)
  s = patt4.sub('', s)

  return unidecode(s)
  
spark.udf.register('func6', str_normalize)
```

```python
def process_38(s):
  d = {'1.0': 1.0, '0.59': 0.59, '0': 0.0, 'DN': np.nan, 'TN': np.nan, '1': 1.0, '3.0': 3.0}
  for k, v in d.items():
    if s == k:
      return v
  else:
    return np.nan
spark.udf.register('func7', process_38)
```

```python
def process_62(s):
  d = {"I": 1, "II": 2, "III": 3, "IV": 4, "V": 5, "Ngoài quốc doanh Quận 7": np.nan}
  for k, v in d.items():
    if s == k:
      return v
  else:
    return np.nan

spark.udf.register('func8', process_62)
```

```python
def process_47(s):
  d = {"Zezo": 0, "One": 1, "Two": 2, "Three": 3, "Four": 4}
  for k, v in d.items():
    if s == k:
      return v
  else:
    return np.nan

spark.udf.register('func9', process_47)
```

```python
def process_2lr_45(s):
  try:
    return s[:2]
  except:
    return np.nan
    
spark.udf.register('func10', process_2lr_45)
```

```python
def process_3lr_45(s):
  try:
    return s[:4]
  except:
    return np.nan
    
spark.udf.register('func11', process_3lr_45)
```

```python
def normalize_gender(s):
  s = str(s).lower()
  if s=='male' or s=='female':
    return s
  return np.nan
  
spark.udf.register('func12', normalize_gender)
```

```python
def combine_gender(a, b):
  if a=='male':
    if b=='male': return 'male'
    if b=='female': return np.nan
    return a
  if a=='female':
    if b=='female': return 'female'
    if b=='male': return np.nan
    return a
  if b=='male':
    if a=='male': return 'male'
    if a=='female': return np.nan
    return b
  if b=='female':
    if a=='female': return 'female'
    if a=='male': return np.nan
    return b
  return np.nan

sum_genders = udf(combine_gender, StringType())
```

```python
def process_locations(s):
  if str(s)=='0.0' or str(s)=='None': return np.nan
  return s

spark.udf.register('func13', process_locations)
```

```python
unicode_cols = ['Field_18', 'maCv', 'diaChi', 'Field_46', 'Field_48', 'Field_49', 'Field_56', 'Field_61', 'homeTownCity', 
                'homeTownName', 'currentLocationCity', 'currentLocationName', 'currentLocationState', 'homeTownState']
object_cols = (unicode_cols + 
               [f'Field_{str(i)}' for i in '4 12 36 38 47 62 45 54 55 65 66 68'.split()] +
               ['data.basic_info.locale', 'currentLocationCountry', 'homeTownCountry', 'brief'])

def process_categorical_cols(df):
  df = df.withColumn('diaChi', split('diaChi', ',')).withColumn('diaChi', element_at('diaChi', -1))
  df = df.withColumn('Field_48', split('Field_48', ',')).withColumn('Field_48', element_at('Field_48', -1))
  df = df.withColumn('Field_49', split('Field_49', ',')).withColumn('Field_49', element_at('Field_49', -1))

  for col in unicode_cols:
    df = df.withColumn(col, expr(f"func6({col})"))


  df = df.withColumn('Field_38', expr("func7(Field_38)"))
  df = df.withColumn('Field_38', df['Field_38'].cast(DoubleType()))
  df = df.withColumn('Field_62', expr("func8(Field_62)"))
  df = df.withColumn('Field_62', df['Field_62'].cast(DoubleType()))
  df = df.withColumn('Field_47', expr("func9(Field_47)"))
  df = df.withColumn('Field_47', df['Field_47'].cast(DoubleType()))


  # new columns
  df = df.withColumn('F48_49', df.Field_48 == df.Field_49)
  df = df.withColumn('is_homeTown_diaChi', df['homeTownCity'] == df['diaChi'])
  df = df.withColumn('is_homeTown_current_city', df['homeTownCity'] == df['currentLocationCity'])
  df = df.withColumn('is_homeTown_current_state', df['homeTownState'] == df['currentLocationState'])

  df = df.withColumn('Field_45_2lr', expr("func10(Field_45)"))
  df = df.withColumn('Field_45_TP_55', df['Field_45_2lr']==df['Field_55'])
  
  df = df.withColumn('Field_45_Q', expr("func11(Field_45)"))
  df = df.withColumn('gioiTinh', expr("func12(gioiTinh)"))
  df = df.withColumn('info_social_sex', expr("func12(info_social_sex)"))
  df = df.withColumn('gender', sum_genders('gioiTinh', 'info_social_sex'))

  df = df.withColumn('currentLocationLocationId', expr("func13(currentLocationLocationId)"))
  df = df.withColumn('homeTownLocationId', expr("func13(homeTownLocationId)"))
  df = df.withColumn('currentLocationLatitude', expr("func13(currentLocationLatitude)"))
  df = df.withColumn('currentLocationLongitude', expr("func13(currentLocationLongitude)"))
  df = df.withColumn('homeTownLatitude', expr("func13(homeTownLatitude)"))
  df = df.withColumn('homeTownLongitude', expr("func13(homeTownLongitude)"))

  columns_to_drop = ['Field_45_2lr', 'info_social_sex', 'gioiTinh']
  df = df.drop(*columns_to_drop)

  return df
```

```python
def process_27_28(s):
  if str(s)=='0.0' or str(s)=='None': return np.nan
  return s

spark.udf.register('func14', process_27_28)
```

```python
def process_18(s):
  try:
    if s.isdigit():
      return True
    else:
      return False
  except:
    return False

spark.udf.register('func15', process_18)
```

```python
def process_deltas(a, b):
  try:
    return a-b
  except:
    return np.nan

diff_a_b = udf(process_deltas, DoubleType())
```

```python
def process_others(df):
  # new cols
  df = df.withColumn('Field_27', expr("func14(Field_27)"))
  df = df.withColumn('Field_28', expr("func14(Field_28)"))
  df = df.withColumn('F18_isdigit', expr("func15(Field_18)"))
  
  # delta cols
  for i, j in [(20, 27), (28, 27), (39, 41), (41, 42), (50, 51), (51, 53)]:
    df = df.withColumn(f'F{i}_{j}_delta', diff_a_b(df[f'Field_{j}'], df[f'Field_{i}']))

  df = df.drop(*ignore_columns)

  return df
```
